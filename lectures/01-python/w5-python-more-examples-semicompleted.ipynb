## Exercise 2: Pascal's triangle

Pascal's triangle is created such that each layer has 1 more element than the previous, with `1`s on the side and in the interior, the numbers are the sum of the two above it, e.g.,:
```
            1
          1   1
        1   2   1
      1   3   3   1
    1   4   6   4   1
  1   5   10  10  5   1
```

Write a function to return the first `n` rows of Pascal's triangle.  The return should be a list of length `n`, with each element itself a list containing the numbers for that row.

<span class="fa fa-star" /> If you want to add complexity, write a function to print out Pascal's triangle with proper formatting, so the numbers in each row are centered between the ones in the row above

````python
#Exercise 2
from scipy.special import binom
rows = input("Choose how many rows to display: \n")
triangle = [[1.0]]
for i in range(1,int(rows)):
    row = []
    for j in range(i+1):
        row.append(binom(i,j))
    triangle.append(row)
for i in triangle:
    print(i)
````

## Exercise 3: Panagrams

A _panagram_ is a sentence that includes all 26 letters of the alphabet, e.g., "_The quick brown fox jumps over the lazy dog_."

Write a function that takes as an argument a sentence and returns `True` or `False`, indicating whether the sentence is a panagram.

````python
def panagram(sentence):
    """Evaluates if the sentence is a panagram
       parameters: sentence a string
       ----------
       returns: true or false 
    """
    sentence = sentence.lower()
    lst = 'abcdefghijklmnopqrstuvwxyz'
    count = 0
    for i in lst:
        if i in sentence:
            count += 1
        else:
            continue
    if count == 26:
        return True
    else:
        return False
````

## Exercise 5: Calendar events

We want to keep a schedule of events.  We will do this by creating a class called `Day`.  It is sketched out below.  A `Day` holds a list of events and has methods that allow you to add an delete events.  Our events will be instances of a class `Event`, which holds the time, location, and description of the event.

Finally, we can keep track of a list of all the `Day`s for which we have events to make our schedule.

Fill in these classes and write some code to demonstrate their use:

  * Create a full week of days in your calendar
  * Add an event every day at noon called "lunch"
  * Randomly add some other events to fill out your calendar
  * Write some code that tells you the start time of your first meeting and the end time of your last meeting (this is the length of your work day)

````python
class Event:
    """a single event in our calendar"""
    def __init__(self,name, time, location=None, duration=None):
        self.name = name
        self.time = time # (hh,mm) a tuple
        self.location = location
        self.duration = duration
        
class Day(object):
    """a single day keeping track of the events scheduled"""
    def __init__(self,month, day, year):
        # store the month, day, and year as data in the class
        self.month = month
        self.day   = day
        self.year  = year
        
        # keep track of the events
        self.events = []
    
    def add_event(self, event):
        """ Adds an event to the list of events"""
        
        self.events.append(event)
     
    
    def delete_event(self,name):
        for i in self.events:
            if i.name == name:
                self.remove(i)
               
            
    def __eq__(self, other):
        return (self.month == other.month and 
                self.day == other.day and 
                self.year == other.year)
    
````

````python
#Create a full week of days in your calendar
week=[]
for i in range(7):
    # Just an easy example, we should use calendar to do this as general as possible  
    week.append(Day(3,i+2,2020)) # Week starting on day 2nd of march

#Add an event every day at noon called "lunch"  
for i in week:
    event=Event("lunch", (12,0), location=None, duration=1)
    i.add_event(event)
    
#Randomly add some other events to fill out your calendar  
event=Event("phy456",(3,0),"b131",1)
week[0].add_event(event)

event=Event("colloquium",(4,15),"Harriman",1.5)
week[1].add_event(event)

event=Event("climbing",(6,0),"Gym",4)
week[2].add_event(event)

event=Event("phy598",(4,0),"Frey308",1.5)
week[3].add_event(event)
    
#Write some code that tells you the start time of your
#first meeting and the end time of your last meeting  
def work_day(day):
    """args: day (a single day of the Day class)
       output: time of first meeting and time of last meeting"""
    
    mintime=0
    maxtime=24
    for i in day.events:
        time = i.time[0] + i.time[1]/60 #convert to hours
        if time > mintime:
            mintime = time
            start=day.events.index(i)
        if time < maxtime:
            maxtime=time
            end=day.events.index(i)
    
    print("Your first meeting on {}/{}/{} is {} at {}h:{}min"
          .format(day.month,day.day,day.year,day.events[start].name,
                 day.events[start].time[0],day.events[start].time[1]))
    
    print("Your last meeting on {}/{}/{} is {} at {}h:{}min"
          .format(day.month,day.day,day.year,day.events[end].name,
                 day.events[end].time[0],day.events[end].time[1]))
````
